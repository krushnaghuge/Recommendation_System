```python
import pandas as pd
import numpy as np
import os
```

```python
BOOK=pd.read_csv('book.csv',encoding=('ISO-8859-1'))
BOOK
```

```python
book1 = BOOK.iloc[:,1:4]
book1.head(5)
```

```python
len(book1['User.ID'].unique())
```

```python
len(book1['Book.Title'].unique())
```

```python
book1.rename({'User.ID':'userid','Book.Title':'title','Book.Rating':'rating'},axis=1,inplace=True)
```

```python
book1
```

```python
book= book1.drop_duplicates()
book
```

```python
book.head()
book.shape
```

```python
# Apply Poviting
df = book.pivot(index='title',columns='userid',values='rating').reset_index(drop=True) 
df
```

```python
df.index = book1.title.unique()
```

```python
df.isna().sum()
```

```python
#Impute those NaNs with 0 values
df.fillna(0, inplace=True)
df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# inverse frequencies as many books are not rated 
books_sim = pairwise_distances(df.values,metric='cosine') 
```

```python
books_sim
```

```python
#Store the results in a dataframe
df_sim = pd.DataFrame(books_sim)
#Set the index and column names to title
df_sim.index = book1.title.unique()
df_sim.columns = book1.title.unique()
df_sim
```

```python
np.fill_diagonal(books_sim, 0)
df_sim
```

```python
#Highest rated book by most of the users
df_sim.idxmax(axis=1)[0:5]
```
